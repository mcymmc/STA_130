1. The difference between the two is that Simple Linear Regression (SLR) only has one independent variable, while Multiple Linear Regression (MLR) has multiple. This means SLR is a method that predicts an outcome by observing the relationship between one variable and the predictor. In contrast, MLR is a method that predicts an outcome by observing multiple relationships between multiple variables and the predictor. The latter has the benefit of providing a more detailed understanding of what causes Y to change.

2. The difference is that continuous variables show us how the independant variables affect the dependant variable as it changes within a range. But with indicator variables, the data is not continuous and therefore we can't see the affects it has across a range, rather we see the changes of the independant variable across each category. 

3. When an indicator variable is added alongside a continuous variable, the SLR becomes an MLR. Since an indicator variable is categorical, the model will change into multiple lines of the same slope to represent each category. Each line represents the line of each category added with the effects of the continuous variable.

4. When an interaction is added between a continuous and indicator variable (multiplying the continuous and indicator variable together and adding it to the linear regression), the slopes of the lines per category can now be different. What we're doing is allowing the effects of the continous variable vary through each category. Earlier, for question 3, the model changes to multiple lines, but of the same slope. Now, there are multiple lines of different slope.  

5. In an MLR of only indicator variables, it captures how the outcome differs by category, this time, without the effects of any continous variables. The indicator itself carries a value that adjusts the intercept value (from the reference). There can be multiple indicator variables, each being its own category and its own line. However, since there is no continuous variable, the lines are actually flat, with slope 0. The way the non-binary categories encoding works is that we use multiple instances of 0s and 1s to represent multiple categories. So for example, if the categories are red, green, yellow, orange, and blue, with blue being the reference category, we would have a combination of four binary values. It might look like 0,1,0,0. Let red represent the first value, green the second, yellow the third and orange the fourth. when the indicator is "1", it means that it is that color, when it is "0", it means it's not. So going back to our earlier example, 0,1,0,0 would mean, not red, yes green, not yellow, not orange, which essentially means the category is green. Note that there can only be one instance of "1" in the combination because there cannot be a category that represents two categories. When the combination is 0,0,0,0, it means none of the four colours, so it is the reference colour, blue.
